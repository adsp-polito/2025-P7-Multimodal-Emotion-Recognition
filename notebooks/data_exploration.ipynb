```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

def emotion_percent(dataset):
    # از متادیتا استفاده می‌کنیم، نه audio
    labels = [item["emo"] for item in dataset.data_list]
    total = len(labels)
    counts = Counter(labels)
    perc = {emo: (cnt / total) * 100 for emo, cnt in counts.items()}
    return perc

# محاسبه درصدها
esd_pct = emotion_percent(esd_train)
iemocap_pct = emotion_percent(iemocap_train)

# لیست همه‌ی emotionها
all_emotions = sorted(set(list(esd_pct.keys()) + list(iemocap_pct.keys())))

esd_values = [esd_pct.get(e, 0) for e in all_emotions]
iemocap_values = [iemocap_pct.get(e, 0) for e in all_emotions]

x = np.arange(len(all_emotions))
width = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x - width/2, esd_values, width, label='ESD')
plt.bar(x + width/2, iemocap_values, width, label='IEMOCAP')

plt.xticks(x, all_emotions, rotation=45)
plt.ylabel("Percent %")
plt.title("Normalized Emotion Distribution (ESD vs IEMOCAP)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def build_length_df(dataset, dataset_name, max_samples_per_emo=2000):
    """
    Build a DataFrame containing:
        - dataset name
        - emotion label
        - utterance length (in seconds)

    max_samples_per_emo: limits the number of samples per emotion to keep
    the violin plot lighter and faster (optional but recommended).
    """

    rows = []
    per_emo_counts = {}

    for item in dataset.data_list:
        emo = item["emo"]
        length = float(item["length"])

        # Limit number of samples per emotion to avoid overplotting
        count = per_emo_counts.get(emo, 0)
        if count >= max_samples_per_emo:
            continue
        per_emo_counts[emo] = count + 1

        rows.append({
            "dataset": dataset_name,
            "emotion": emo,
            "length": length,
        })

    return pd.DataFrame(rows)

# Build DataFrame for both datasets
df_esd = build_length_df(esd_train, "ESD")
df_iemocap = build_length_df(iemocap_train, "IEMOCAP")

# Merge into one table
df_all = pd.concat([df_esd, df_iemocap], ignore_index=True)

# Violin plot
plt.figure(figsize=(10, 5))
sns.violinplot(
    data=df_all,
    x="emotion",
    y="length",
    hue="dataset",
    split=True
)

plt.ylabel("Utterance length (seconds)")
plt.title("Length distribution per emotion (ESD vs IEMOCAP)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def extract_speaker_from_key(key: str) -> str:
    """
    Extract speaker ID from IEMOCAP key.

    Example:
        key = 'iemocap-Ses02F_impro07_F001'
        speaker ID = 'F001'

    If the format does not match, returns 'Unknown'.
    """

    if key.startswith("iemocap-"):
        base = key[len("iemocap-"):]  # Remove prefix → 'Ses02F_impro07_F001'
        parts = base.split("_")
        if len(parts) >= 3:
            return parts[-1]  # Last part is speaker ID (e.g., 'F001')
    return "Unknown"

# Build the emotion × speaker table
rows = []
for item in iemocap_train.data_list:
    emo = item["emo"]
    speaker = extract_speaker_from_key(item["key"])
    rows.append({"speaker": speaker, "emotion": emo})

df_spk = pd.DataFrame(rows)

# Pivot table for heatmap
pivot = pd.crosstab(df_spk["speaker"], df_spk["emotion"])

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot, cmap="viridis")
plt.title("IEMOCAP: Emotion × Speaker (train)")
plt.ylabel("Speaker")
plt.xlabel("Emotion")
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Helper functions
# ----------------------------

def get_labels(dataset):
    """
    Extract emotion labels.
    Use 'label' if available, otherwise fallback to metadata field 'emo'.
    """
    return [item.get("label", item["emo"]) for item in dataset.data_list]


def get_lengths(dataset):
    """
    Extract audio lengths (seconds) from each item.
    """
    return [item["length"] for item in dataset.data_list]


def summarize_dataset(name, dataset):
    """
    Print dataset summary:
    - number of samples
    - emotion distribution
    - length statistics
    Returns:
    - label_counts (Counter)
    - lengths (list of floats)
    """
    labels = get_labels(dataset)
    lengths = get_lengths(dataset)

    print(f"===== {name} =====")
    print(f"Total samples: {len(dataset)}")

    # Emotion counts
    label_counts = Counter(labels)
    print("\nEmotion distribution:")
    for emo, cnt in label_counts.items():
        print(f"{emo:10s} {cnt:5d}")

    # Length stats
    print("\nLength stats:")
    print("min:", min(lengths))
    print("max:", max(lengths))
    print("mean:", np.mean(lengths))
    print("std:", np.std(lengths))
    print("-----------------------------------")

    return label_counts, lengths


# ----------------------------
# Summaries for Both Datasets
# ----------------------------
esd_counts, esd_lengths = summarize_dataset("ESD (Train)", esd_train)
iemocap_counts, iemocap_lengths = summarize_dataset("IEMOCAP (Train)", iemocap_train)


# ----------------------------
# Emotion Bar Charts (Side-by-side)
# ----------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(esd_counts.keys(), esd_counts.values())
plt.title("ESD Emotion Distribution")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.bar(iemocap_counts.keys(), iemocap_counts.values())
plt.title("IEMOCAP Emotion Distribution")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


# ----------------------------
# Length Histogram (Side-by-side)
# ----------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(esd_lengths, bins=40)
plt.title("ESD Length Distribution")

plt.subplot(1,2,2)
plt.hist(iemocap_lengths, bins=40)
plt.title("IEMOCAP Length Distribution")

plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import numpy as np

def summarize_categorical(name, data_list, attr):
    values = [item.get(attr, "Unknown") for item in data_list]
    counts = Counter(values)
    total = len(values)
    print(f"=== {name} – {attr} distribution ===")
    for v, c in counts.items():
        print(f"{str(v):10s}: {c:6d}  ({c/total*100:5.1f}%)")
    print()

def summarize_length(name, data_list):
    lengths = np.array([float(item.get("length", 0.0)) for item in data_list])
    print(f"=== {name} – length stats (seconds) ===")
    print("count :", len(lengths))
    print("min   :", lengths.min())
    print("max   :", lengths.max())
    print("mean  :", lengths.mean())
    print("std   :", lengths.std())
    print()
```

```python
import os
import json
import random
from collections import Counter, defaultdict

import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

# Make runs reproducible
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# ============================================================
# 0) Small helpers to peek into samples
# ============================================================

def get_labels(dataset):
    """Return a list of standardized emotion labels from the dataset."""
    return [item["emo"] for item in dataset.data_list]

def get_lengths(dataset):
    """Return a list of utterance lengths (in seconds) from the dataset."""
    return [float(item["length"]) for item in dataset.data_list]

def get_attr_list(dataset, key, default=None):
    """
    Generic helper to extract a list of a given metadata key from the dataset.
    If a sample does not contain the key, use the given default (e.g. 'Unknown').
    """
    values = []
    for item in dataset.data_list:
        if key in item and item[key] not in [None, ""]:
            values.append(item[key])
        else:
            values.append(default)
    return values


# ============================================================
# 1) Label Standardization & Distribution
# ============================================================

def load_label_map(meta_data_dir, dataset_name):
    """
    Load raw->standardized label map from label_map.json for a given dataset.
    This is exactly the mapping EmoDataset already used internally.
    """
    lm_path = os.path.join(meta_data_dir, dataset_name, "label_map.json")
    if not os.path.exists(lm_path):
        print(f"[WARNING] No label_map.json found for {dataset_name} at {lm_path}")
        return {}
    with open(lm_path, "r") as f:
        label_map = json.load(f)
    return label_map

def standardized_label_stats(dataset, dataset_name):
    """
    Compute counts + percentages for standardized labels in a dataset.
    Also identify rare classes (e.g. classes with < 5% of samples).
    """
    labels = get_labels(dataset)
    counts = Counter(labels)
    total = len(labels)

    print(f"\n===== Standardized label stats for {dataset_name} =====")
    rare_threshold = 0.05  # 5%
    rare_classes = []

    stats = {}
    for lab, cnt in counts.items():
        pct = cnt / total
        stats[lab] = {"count": cnt, "percent": pct}
        rare_flag = "  (RARE)" if pct < rare_threshold else ""
        if pct < rare_threshold:
            rare_classes.append(lab)
        print(f"{lab:10s}  count = {cnt:5d}  pct = {pct*100:6.2f}%{rare_flag}")

    if rare_classes:
        print(f"--> Rare classes (<5% of samples): {rare_classes}")
    else:
        print("--> No rare classes under 5% threshold.")

    return stats, rare_classes


# ============================================================
# 2) Sensitive Attributes (age, gender, speaker, session)
# ============================================================

def summarize_sensitive_attribute(dataset, dataset_name, key, bins=None):
    """
    Summarize a sensitive attribute (e.g., gender, age) for a dataset.
    - If bins is None: treat as categorical.
    - If bins is a list: treat as numeric (e.g., age) and bin values.
    Returns: dict with counts, missing_rate, and (optionally) binned counts.
    """
    values_raw = []
    missing = 0
    for item in dataset.data_list:
        if key in item and item[key] not in [None, ""]:
            values_raw.append(item[key])
        else:
            missing += 1

    total = len(dataset)
    missing_rate = missing / total

    print(f"\n===== Sensitive attribute: {key} ({dataset_name}) =====")
    print(f"Total samples: {total}")
    print(f"Missing / Unknown: {missing} ({missing_rate*100:.2f}%)")

    result = {
        "total": total,
        "missing": missing,
        "missing_rate": missing_rate,
    }

    # Categorical attribute (e.g., 'gender')
    if bins is None:
        if not values_raw:
            print("No values present for this attribute.")
            result["counts"] = {}
            return result

        counts = Counter(values_raw)
        for val, cnt in counts.items():
            print(f"{val:10s}  {cnt:5d}  ({cnt/total*100:6.2f}%)")
        result["counts"] = counts
        return result

    # Numeric attribute with bins (e.g., 'age')
    else:
        # Convert to numeric where possible
        numeric_vals = []
        for v in values_raw:
            try:
                numeric_vals.append(float(v))
            except Exception:
                # Skip non-numeric values
                continue

        if not numeric_vals:
            print("No valid numeric values for this attribute.")
            result["bins"] = {}
            return result

        # Example bins: [0, 18, 25, 35, 50, 100]
        bin_edges = bins
        bin_labels = []
        for i in range(len(bin_edges) - 1):
            bin_labels.append(f"[{bin_edges[i]}, {bin_edges[i+1]})")

        hist, _ = np.histogram(numeric_vals, bins=bin_edges)
        bin_counts = {bin_labels[i]: int(hist[i]) for i in range(len(hist))}

        print("Age-bin distribution:")
        for label in bin_labels:
            cnt = bin_counts[label]
            print(f"{label:12s} {cnt:5d} ({cnt/total*100:6.2f}%)")

        result["bins"] = bin_counts
        return result

def emotion_by_gender(dataset, dataset_name):
    """
    Compute emotion distribution by gender:
    count and percentage for each (gender, emotion) pair.
    """
    genders = get_attr_list(dataset, "gender", default="Unknown")
    emos = get_labels(dataset)

    joint_counts = defaultdict(Counter)
    for g, e in zip(genders, emos):
        joint_counts[g][e] += 1

    print(f"\n===== Emotion × Gender ({dataset_name}) =====")
    for g, emo_counts in joint_counts.items():
        total_g = sum(emo_counts.values())
        print(f"\nGender: {g} (n = {total_g})")
        for emo, cnt in emo_counts.items():
            print(f"  {emo:10s}  {cnt:5d}  ({cnt/total_g*100:6.2f}%)")

    return joint_counts

def emotion_by_age(dataset, dataset_name, bins=None):
    """
    Compute emotion distribution by age bins.
    If bins is None, a default binning is used.
    """
    if bins is None:
        # Simple, explainable age bins (young / adult / mid / senior)
        # These are easy to justify in the report.
        bins = [0, 18, 25, 35, 50, 100]

    # First create numeric ages and bin indices
    age_vals = []
    emos = []
    for item in dataset.data_list:
        if "age" in item and item["age"] not in [None, ""]:
            try:
                age_vals.append(float(item["age"]))
                emos.append(item["emo"])
            except Exception:
                continue

    if not age_vals:
        print(f"\n[INFO] No valid ages for emotion×age in {dataset_name}.")
        return {}

    age_vals = np.array(age_vals)
    emos = np.array(emos)

    bin_indices = np.digitize(age_vals, bins, right=False) - 1
    bin_labels = []
    for i in range(len(bins) - 1):
        bin_labels.append(f"[{bins[i]}, {bins[i+1]})")

    joint_counts = defaultdict(Counter)
    for idx, emo in zip(bin_indices, emos):
        if 0 <= idx < len(bin_labels):
            joint_counts[bin_labels[idx]][emo] += 1

    print(f"\n===== Emotion × Age-bin ({dataset_name}) =====")
    for b_label, emo_counts in joint_counts.items():
        total_b = sum(emo_counts.values())
        print(f"\nAge bin: {b_label} (n = {total_b})")
        for emo, cnt in emo_counts.items():
            print(f"  {emo:10s}  {cnt:5d}  ({cnt/total_b*100:6.2f}%)")

    return joint_counts


# ============================================================
# 3) Duration & Sampling-rate Quality Checks
# ============================================================

def duration_stats(dataset, dataset_name):
    """Print basic duration statistics and detect potential outliers."""
    lengths = get_lengths(dataset)
    arr = np.array(lengths, dtype=float)
    print(f"\n===== Duration stats ({dataset_name}) =====")
    print(f"count : {arr.size}")
    print(f"min   : {arr.min():.3f} s")
    print(f"max   : {arr.max():.3f} s")
    print(f"mean  : {arr.mean():.3f} s")
    print(f"median: {np.median(arr):.3f} s")
    print(f"std   : {arr.std():.3f} s")

    # Simple outlier detection using IQR
    q1, q3 = np.percentile(arr, [25, 75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    n_low = (arr < lower).sum()
    n_high = (arr > upper).sum()
    print(f"Potential short outliers (<{lower:.2f}s): {n_low}")
    print(f"Potential long outliers  (>{upper:.2f}s): {n_high}")

    return {
        "min": float(arr.min()),
        "max": float(arr.max()),
        "mean": float(arr.mean()),
        "median": float(np.median(arr)),
        "std": float(arr.std()),
        "n_short_outliers": int(n_low),
        "n_long_outliers": int(n_high),
    }

def sampling_rate_consistency(dataset, dataset_name):
    """
    Check sampling_rate field consistency in metadata.
    (We trust EmoBox metadata instead of opening audio files.)
    """
    sr_values = [item.get("sample_rate", None) for item in dataset.data_list]
    unique = sorted(set(sr_values))
    counts = Counter(sr_values)

    print(f"\n===== Sampling rate consistency ({dataset_name}) =====")
    for sr in unique:
        print(f"{sr}: {counts[sr]} samples")
    return counts


# ============================================================
# 4) Split integrity / leakage & duplicates
# ============================================================

def speaker_overlap(train_ds, val_ds, test_ds, name_prefix=""):
    """
    Check whether there are speakers that appear in more than one split
    (train/val/test). This is important for leakage detection.
    """
    def get_speakers(ds):
        return {item["speaker"] for item in ds.data_list if "speaker" in item}

    train_spk = get_speakers(train_ds)
    val_spk   = get_speakers(val_ds)
    test_spk  = get_speakers(test_ds)

    print(f"\n===== Speaker overlap {name_prefix} =====")
    print(f"#train speakers: {len(train_spk)}")
    print(f"#val   speakers: {len(val_spk)}")
    print(f"#test  speakers: {len(test_spk)}")

    print("train ∩ val :", len(train_spk & val_spk))
    print("train ∩ test:", len(train_spk & test_spk))
    print("val   ∩ test:", len(val_spk & test_spk))

    return {
        "train_only": list(train_spk - val_spk - test_spk),
        "val_only":   list(val_spk - train_spk - test_spk),
        "test_only":  list(test_spk - train_spk - val_spk),
        "train_val":  list(train_spk & val_spk),
        "train_test": list(train_spk & test_spk),
        "val_test":   list(val_spk & test_spk),
    }

def detect_duplicate_wavs(*datasets, name_prefix=""):
    """
    Detect duplicate audio file paths across multiple splits.
    Uses the 'wav' metadata field (relative paths).
    """
    path_to_splits = defaultdict(set)
    for split_name, ds in datasets:
        for item in ds.data_list:
            wav = item.get("wav", None)
            if wav is not None:
                path_to_splits[wav].add(split_name)

    dup_across_splits = {p: s for p, s in path_to_splits.items() if len(s) > 1}

    print(f"\n===== Duplicate wav paths across splits {name_prefix} =====")
    print(f"Total unique wav paths: {len(path_to_splits)}")
    print(f"Duplicates across splits: {len(dup_across_splits)}")
    return dup_across_splits


# ============================================================
# 5) Cross-dataset comparability for standardized labels
# ============================================================

def compare_label_spaces(stats_esd, stats_iemocap):
    """
    Compare which standardized labels are shared between ESD and IEMOCAP,
    and which are unique to each dataset.
    """
    labels_esd = set(stats_esd.keys())
    labels_iem = set(stats_iemocap.keys())

    shared = labels_esd & labels_iem
    only_esd = labels_esd - labels_iem
    only_iem = labels_iem - labels_esd

    print("\n===== Cross-dataset label sets =====")
    print("Shared labels:        ", sorted(shared))
    print("Only in ESD:          ", sorted(only_esd))
    print("Only in IEMOCAP:      ", sorted(only_iem))
    return shared, only_esd, only_iem

def plot_standardized_label_comparison(stats_esd, stats_iemocap):
    """
    Make a side-by-side bar chart of percentage distributions
    (using the same standardized label set).
    """
    all_labels = sorted(set(stats_esd.keys()) | set(stats_iemocap.keys()))
    esd_pct = [stats_esd.get(l, {"percent": 0})["percent"] * 100 for l in all_labels]
    iem_pct = [stats_iemocap.get(l, {"percent": 0})["percent"] * 100 for l in all_labels]

    x = np.arange(len(all_labels))
    width = 0.35

    plt.figure(figsize=(10, 4))
    plt.bar(x - width/2, esd_pct, width, label="ESD")
    plt.bar(x + width/2, iem_pct, width, label="IEMOCAP")
    plt.xticks(x, all_labels, rotation=45)
    plt.ylabel("Percent %")
    plt.title("Standardized Emotion Distribution (ESD vs IEMOCAP)")
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================================================
# 6) Generic metadata quality scan
# ============================================================

def metadata_quality_scan(dataset, dataset_name):
    """
    Automatically scan all keys in the JSON metadata and:
    - compute missing rate per key
    - for numeric-looking keys, print min/max
    """
    print(f"\n===== Metadata quality scan ({dataset_name}) =====")
    all_keys = set()
    for item in dataset.data_list:
        all_keys.update(item.keys())

    total = len(dataset)
    summary = {}

    for k in sorted(all_keys):
        present_vals = []
        missing = 0
        for item in dataset.data_list:
            if k in item and item[k] not in [None, ""]:
                present_vals.append(item[k])
            else:
                missing += 1
        missing_rate = missing / total

        info = {"missing": missing, "missing_rate": missing_rate}
        # Try to see if values look numeric
        numeric_vals = []
        for v in present_vals:
            try:
                numeric_vals.append(float(v))
            except Exception:
                numeric_vals = []
                break

        if numeric_vals:
            arr = np.array(numeric_vals)
            info["min"] = float(arr.min())
            info["max"] = float(arr.max())

        summary[k] = info
        print(f"{k:12s} missing: {missing:5d} ({missing_rate*100:6.2f}%)",
              f"  numeric-range: [{info.get('min', 'NA')}, {info.get('max', 'NA')}]")

    return summary


# ============================================================
# 7) Example usage on ESD & IEMOCAP
# ============================================================

# --- 7.1 Label maps and standardized label distributions
esd_label_map     = load_label_map(meta_data_dir, "esd")
iemocap_label_map = load_label_map(meta_data_dir, "iemocap")

print("\nRaw → standardized label map (ESD):")
print(esd_label_map)
print("\nRaw → standardized label map (IEMOCAP):")
print(iemocap_label_map)

esd_label_stats, esd_rare = standardized_label_stats(esd_train, "ESD train")
iem_label_stats, iem_rare = standardized_label_stats(iemocap_train, "IEMOCAP train")

shared_labels, only_esd, only_iem = compare_label_spaces(esd_label_stats, iem_label_stats)
plot_standardized_label_comparison(esd_label_stats, iem_label_stats)

# --- 7.2 Sensitive attributes: gender / age
gender_esd   = summarize_sensitive_attribute(esd_train, "ESD train", "gender")
gender_iemoc = summarize_sensitive_attribute(iemocap_train, "IEMOCAP train", "gender")

# Example numeric bins for age: [0,18), [18,25), [25,35), [35,50), [50,100)
age_bins = [0, 18, 25, 35, 50, 100]
age_esd   = summarize_sensitive_attribute(esd_train, "ESD train", "age",  bins=age_bins)
age_iemoc = summarize_sensitive_attribute(iemocap_train, "IEMOCAP train", "age", bins=age_bins)

# Emotion × gender / age (only meaningful if attributes exist)
emo_gender_esd   = emotion_by_gender(esd_train, "ESD train")
emo_gender_iemoc = emotion_by_gender(iemocap_train, "IEMOCAP train")

emo_age_esd   = emotion_by_age(esd_train, "ESD train", bins=age_bins)
emo_age_iemoc = emotion_by_age(iemocap_train, "IEMOCAP train", bins=age_bins)

# --- 7.3 Duration & sampling rate checks
dur_esd   = duration_stats(esd_train, "ESD train")
dur_iemoc = duration_stats(iemocap_train, "IEMOCAP train")

sr_esd   = sampling_rate_consistency(esd_train, "ESD train")
sr_iemoc = sampling_rate_consistency(iemocap_train, "IEMOCAP train")

# --- 7.4 Split integrity / duplicates
esd_overlap = speaker_overlap(esd_train, esd_val, esd_test, name_prefix="(ESD)")
iem_overlap = speaker_overlap(iemocap_train, iemocap_val, iemocap_test, name_prefix="(IEMOCAP)")

esd_dups = detect_duplicate_wavs(
    ("train", esd_train), ("val", esd_val), ("test", esd_test),
    name_prefix="(ESD)"
)
iem_dups = detect_duplicate_wavs(
    ("train", iemocap_train), ("val", iemocap_val), ("test", iemocap_test),
    name_prefix="(IEMOCAP)"
)

# --- 7.5 Metadata quality quick scan
meta_esd   = metadata_quality_scan(esd_train, "ESD train")
meta_iemoc = metadata_quality_scan(iemocap_train, "IEMOCAP train")

# ------------------------------------------------------------
# At this point you already have:
# - Label standardization report + rare classes
# - Sensitive attribute coverage and emotion×gender/age
# - Duration & sampling-rate diagnostics
# - Split leakage / duplicates
# - Cross-dataset label comparison + plot
# - Automatic metadata quality summary
# Everything can be saved later into a JSON summary for reproducibility.
# ------------------------------------------------------------
```
